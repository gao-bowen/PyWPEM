# 批量计算

```python
import os
import shutil
import pandas as pd
import numpy as np

# -------------------------------------------------
# Initial lattice constants (seed value for fitting)
# -------------------------------------------------
Lattice_constants = [[
    2.87549,
    2.87549,
    14.02056,
    90.0,
    90.0,
    120.0
]]

# Dictionary to store results
results = {}

# Number of samples
num_samples = 5 ##############

# -------------------------------------------------
# Main loop over all XRD files
# -------------------------------------------------
for i in range(num_samples):

    # Sample index (start from 1)
    idx = i + 1

    # Input XRD file path
    file_name = f'data/{idx}.xlsx'

    # Temporary working directory for each sample
    tmp_dir = f'tmp/{idx}'
    os.makedirs(tmp_dir, exist_ok=True)

    try:
        # -------------------------------------------------
        # Step 1: Copy raw intensity file to temp folder
        # -------------------------------------------------
        tmp_intensity = os.path.join(tmp_dir, "intensity.xlsx")
        shutil.copyfile(file_name, tmp_intensity)

        # Copy required auxiliary file
        shutil.copyfile(
            "peak0.csv",
            os.path.join(tmp_dir, "peak0.csv")
        )

        # -------------------------------------------------
        # Step 2: Load intensity data
        # -------------------------------------------------
        intensity_excel = pd.read_excel(
            tmp_intensity,
            header=None
        )

        # -------------------------------------------------
        # Step 3: Background fitting
        # -------------------------------------------------
        var = WPEM.BackgroundFit(
            intensity_excel,
            lowAngleRange=17,
            LFctg=0.0,
            poly_n=8,
            bac_split=10,
            bac_num=80,
            work_dir=tmp_dir
        )

        # -------------------------------------------------
        # Step 4: Output file paths generated by WPEM
        # -------------------------------------------------
        no_bac_file = os.path.join(
            tmp_dir,
            "ConvertedDocuments",
            "no_bac_intensity.csv"
        )

        bac_file = os.path.join(
            tmp_dir,
            "ConvertedDocuments",
            "bac.csv"
        )

        # -------------------------------------------------
        # Step 5: XRD fitting
        # -------------------------------------------------
        wavelength = [1.540593, 1.544414]

        _, _lattice = WPEM.XRDfit(
            wavelength,
            var,
            Lattice_constants,
            no_bac_file,
            tmp_intensity,
            bac_file,
            Ave_Waves=True,
            subset_number=6,
            low_bound=15,
            up_bound=60,
            bta=0.85,
            iter_max=3, ##############
            asy_C=0,
            InitializationEpoch=0,
            work_dir=tmp_dir
        )

        # Extract optimized lattice parameters
        lattice = _lattice[0]

        # Update initial lattice for next iteration (sequential refinement)
        Lattice_constants = [lattice.tolist()]

        # Store result
        results[file_name] = lattice.tolist()

        print(f"Done: {file_name}")

    except Exception as e:

        print(f"Error {file_name}: {e}")
        results[file_name] = None


# -------------------------------------------------
# Save results to Excel
# -------------------------------------------------

df = pd.DataFrame({
    "File Name": list(results.keys()),
    "Lattice": [
        str(v) if v is not None else None
        for v in results.values()
    ]
})

# Sort by file index for readability
df = df.sort_values("File Name")

# Save to Excel file
output_file = "lattice_results.xlsx"
df.to_excel(output_file, index=False)

print(f"\nResults saved to: {output_file}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast

# =========================
# Load data
# =========================
df = pd.read_excel("lattice_results.xlsx")

# =========================
# Parse lattice string -> list
# =========================
def parse_lattice(x):
    try:
        if isinstance(x, str):
            return ast.literal_eval(x)
        return None
    except:
        return None

df["Lattice_parsed"] = df["Lattice"].apply(parse_lattice)

# =========================
# Extract C-axis (index 2)
# =========================
df["C_axis"] = df["Lattice_parsed"].apply(
    lambda x: x[2] if isinstance(x, list) and len(x) >= 3 else np.nan
)

# =========================
# Sort by file index
# =========================
df["idx"] = df["File Name"].str.extract(r"(\d+)").astype(float)
df = df.sort_values("idx")

# =========================
# Clean data
# =========================
df = df.dropna(subset=["C_axis"])

# =========================
# Nature-style plot
# =========================
plt.figure(figsize=(4.2, 3.2))

plt.plot(
    df["idx"],
    df["C_axis"],
    color="black",
    linewidth=1.5
)

# labels
plt.xlabel("Sample index", fontsize=11)
plt.ylabel("C lattice parameter (Å)", fontsize=11)

# style: Nature-like clean frame
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.tick_params(direction="out", length=4, width=1)

plt.tight_layout()


plt.show()
```
