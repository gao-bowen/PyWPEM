# 安装软件

## 创建新的 conda 环境

### 先打开终端（Mac / Linux / Windows Anaconda Prompt）：

```python
conda create -n myenv python=3.10
```

+ `myenv`：环境名字（可以改成你项目名，比如 xrd-ai）
+ `python=3.10`：建议指定版本

## 激活环境

激活后你会看到前面有 `(myenv)` 标志

```python
conda activate myenv
```

## 安装 Jupyter Notebook

```python
conda install notebook
```

## 安装 PyXplore

```python
pip install PyXplore
```

## 安装 PyXplore 依赖

```python
wget https://raw.githubusercontent.com/Bin-Cao/PyWPEM/main/requirements.txt
```

```python
python -m pip install -r requirements.txt
```

## 启动 Jupyter Notebook

```python
jupyter notebook
```

## 测试

```python
from PyXplore import WPEM
```

```python
!conda info --envs
```
